```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
figdir = './plots'

if not os.path.exists(figdir):
    os.makedirs(figdir)
```

```python
def func(t, s, tau, r):
    phi = ((t - s) % tau)/tau
    if phi <= r/2:
        return 2 * phi/r
    elif phi <= r:
        return 2 - 2*phi/r
    else:
        return 0
```

```python
t = np.arange(0, 3, 0.001)
tau = 1
s = np.arange(0, 1, 0.25)
r = 1
for s_0 in s:
    plt.plot(t, [func(t_0, s_0, tau, r) for t_0 in t], label=f's={s_0}')


plt.xlabel('time (t)')
plt.ylabel('value (k^(t))')
plt.title(f'Various s, tau = {tau}, r = {r}')

plt.legend(loc=1)
plt.savefig(os.path.join(figdir, 's_plot.jpg'), dpi=1000)
plt.show()
```

```python
t = np.arange(0, 3, 0.001)
tau = np.arange(1, 6)
s = 0
r = 1
for tau_0 in tau:
    plt.plot(t, [func(t_0, s, tau_0, r) for t_0 in t], label=f'tau={tau_0}')


plt.xlabel('time (t)')
plt.ylabel('value (k^(t))')
plt.title(f'Various tau, s = {s}, r = {r}')

plt.legend(loc=1)
plt.savefig(os.path.join(figdir, 'tau_plot.jpg'), dpi=1000)
plt.show()
```

```python
t = np.arange(0, 3, 0.001)
tau = 1
s = 0
r = np.arange(0.5, 3, 0.5)
for r_0 in r:
    plt.plot(t, [func(t_0, s, tau, r_0) for t_0 in t], label=f'r={r_0}')


plt.xlabel('time (t)')
plt.ylabel('value (k^(t))')
plt.title(f'Various r, s = {s}, tau = {tau}')

plt.legend(loc=1)
plt.savefig(os.path.join(figdir, 'r_plot.jpg'), dpi=1000)
plt.show()
```
