```python
import torch
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
extensions = ['.txt']
file_dict = {}
for subdir, dirs, files in os.walk('../experiments'):
    for file in files:
        ext = os.path.splitext(file)[-1].lower()
        if file == 'output_200_[512, 512]_0.0001_0.0001_Adam_0.9_[0.2, 0.2].txt':
            path = os.path.join(subdir, file)
            print(path)
            file_dict[path] = {}
            with open(path, 'r') as file:
                lines = file.readlines()
                test_steps = lines[1]
                test_steps = test_steps[test_steps.find('[')+1:test_steps.rfind(']')]
                test_losses = lines[2]
                test_losses = test_losses[test_losses.find('[')+1:test_losses.rfind(']')]
                test_accuracies = lines[3]
                test_accuracies = test_accuracies[test_accuracies.find('[')+1:test_accuracies.rfind(']')]
                train_steps = lines[4]
                train_steps = train_steps[train_steps.find('[')+1:train_steps.rfind(']')]
                train_losses = lines[5]
                train_losses = train_losses[train_losses.find('[')+1:train_losses.rfind(']')]
                train_accuracies = lines[6]
                train_accuracies = train_accuracies[train_accuracies.find('[')+1:train_accuracies.rfind(']')]
                
                test_steps = np.array([int(val.strip()) for val in test_steps.split(',')])
                test_losses= np.array([float(val[8:]) for val in test_losses.split(',') if '.' in val])
                test_accuracies = np.array([float(val[8:]) for val in test_accuracies.split(',') if '.' in val])
                train_steps = np.array([int(val.strip()) for val in train_steps.split(',')])
                train_losses = np.array([float(val[8:]) for val in train_losses.split(',') if '.' in val])
                train_accuracies = np.array([float(val[8:]) for val in train_accuracies.split(',') if '.' in val])
#                 for line in file.readlines():
#                     print(line)
```

```python
_, (loss_axis, accuracy_axis) = plt.subplots(
  nrows=1, ncols=2, figsize=(10, 4)
)

loss_axis.cla()
loss_axis.plot(train_steps[1:], train_losses[1:], label="train loss")
loss_axis.plot(test_steps[1:], test_losses[1:], label="test loss")
loss_axis.legend()
loss_axis.set_title('Train and Test Losses')
loss_axis.set_ylabel('Loss')
loss_axis.set_xlabel('Step')

accuracy_axis.cla()
accuracy_axis.plot(train_steps[1:], train_accuracies[1:], label="train accuracy")
accuracy_axis.plot(test_steps, test_accuracies, label="test accuracy")
accuracy_axis.set_title('Train and Test Accuracies')
accuracy_axis.legend()
accuracy_axis.set_ylabel('Accuracy')
accuracy_axis.set_xlabel('Step')

plt.draw()
plt.ion()
# plt.show()
plt.savefig('../plots/pytorch_mlp.png')
```
